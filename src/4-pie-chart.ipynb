### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Error handling
if len(labels) != len(sizes):
    raise ValueError("Labels and sizes must have the same length.")
if sum(sizes) != 100:
    print("Warning: Sizes do not sum to 100. Adjusting values proportionally.")
    sizes = [s/sum(sizes) * 100 for s in sizes]

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple'])

# Customizations
plt.title("Pie Chart with Error Handling")

plt.show()
```
