### Exercise 6: Histogram Customization

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Error handling
if not all(isinstance(i, (int, float)) for i in data):
    raise ValueError("All elements in data must be numeric.")

# Create histogram
plt.hist(data, bins=5, color='cyan', edgecolor='black')

# Customizations
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram with Error Handling")

plt.show()
```
