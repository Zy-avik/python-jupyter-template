### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Error handling
if len(categories) != len(values):
    raise ValueError("Categories and values must have the same length.")
if not all(isinstance(val, (int, float)) for val in values):
    raise TypeError("Values must be numeric.")
if len(colors) != len(categories):
    print("Warning: Number of colors does not match categories. Using default colors.")

# Create bar plot
plt.bar(categories, values, color=colors)

# Customizations
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot with Error Handling")

plt.show()
```
