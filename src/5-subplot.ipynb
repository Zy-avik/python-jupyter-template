### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
if len(x) == len(y):
    axes[0, 0].plot(x, y, color='b')
    axes[0, 0].set_title("Line Plot")
else:
    axes[0, 0].text(0.5, 0.5, "Error in Data", ha='center')

# Bar plot
if len(categories) == len(values):
    axes[0, 1].bar(categories, values, color=['r', 'g', 'b', 'purple'])
    axes[0, 1].set_title("Bar Chart")
else:
    axes[0, 1].text(0.5, 0.5, "Error in Data", ha='center')

# Scatter plot
if len(scatter_x) == len(scatter_y):
    axes[1, 0].scatter(scatter_x, scatter_y, color='r')
    axes[1, 0].set_title("Scatter Plot")
else:
    axes[1, 0].text(0.5, 0.5, "Error in Data", ha='center')

# Pie chart
if len(labels) == len(sizes) and sum(sizes) == 100:
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
    axes[1, 1].set_title("Pie Chart")
else:
    axes[1, 1].text(0.5, 0.5, "Error in Data", ha='center')

plt.tight_layout()
plt.show()
```
